```python
import pandas as pd

world = pd.read_csv('python-novice-gapminder-data/data/gapminder_all.csv',index_col='country')
lifeexp_data = world.loc[:, 'lifeExp_1952':'lifeExp_2007']
lifeexp_data['continent'] = world['continent']

grouped = lifeexp_data.groupby('continent')
grouped.mean()
lifeexp_mean = grouped.mean()

lifeexp_mean.to_csv('world_lideexp_average.csv')
```
